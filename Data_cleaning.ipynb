```python
movies = pd.read_csv("movies.csv")
#movies.head()

genre_list = movies["genres"].str.split('|').tolist()

genre_dict = {"Action": 1,\
"Adventure": 2,\
"Animation": 3,\
"Children": 4,\
"Comedy": 5,\
"Crime": 6,\
"Documentary": 7,\
"Drama": 8,\
"Fantasy": 9,\
"Film-Noir": 10,\
"Horror": 11,\
"Musical": 12,\
"Mystery": 13,\
"Romance": 14,\
"Sci-Fi": 15,\
"Thriller": 16,\
"War": 17,\
"Western": 18,\
"(no genres listed)": 19}

genre_num = []
for genres in genre_list:
    num_list = []
    for genre in genres:
        if genre in genre_dict:
            num_list.append(str(genre_dict[genre]))
        else:
            num_list.append(str('-1'))
    genre_num.append("|".join(num_list))

movies["genre_no"] = genre_num
movies.head()
```

```python
movies = pd.read_csv("movies.csv")
#movies.head()

genre_list = movies["genres"].str.split('|').tolist()
total_genre_name = []
genre_dict = dict()
for genres in genre_list:
    for genre in genres:
        if genre not in total_genre_name:
            total_genre_name.append(genre)

i = 0
for genre in sorted(total_genre_name, key=str.lower):
    genre_dict[genre] = i
    i += 1

    
genre_num = []
for genres in genre_list:
    num_list = []
    for genre in genres:
        if genre in genre_dict:
            num_list.append(str(genre_dict[genre]))
        else:
            num_list.append(str('-1'))
    genre_num.append("|".join(num_list))

movies["genre_no"] = genre_num
movies.head()
```
